### Fetching Cryptocurrency Data

```python
request_url = "https://min-api.cryptocompare.com/data/all/coinlist"
```

```python
# Submit request and format output
response_data = requests.get(request_url).json()
```

```python
# Submit request and format output
request = Request(request_url)
response = urlopen(request)
url_data = response.read()
data_json = json.loads(url_data)
#data = response_data.read()
```

```python
df = pd.DataFrame(data_json['Data'])
df.head()
```

```python
# Creating a DataFrame
```

### Data Preprocessing

```python
# Traspose the data
df1 = df.T
df1.head()
```

```python
list(df1.columns)
```

```python
# Remove unnecessary columns
df_clean = df1.drop(columns=['BuiltOn',
                 'ContentCreatedOn',
                 'FullName',
                 'Id',
                 'ImageUrl',
                 'PreMinedValue',
                 'SmartContractAddress',
                 'SortOrder',
                 'Sponsored',
                 'TotalCoinsFreeFloat',
                 'Url'])
```

```python
df_clean.head()
```

```python
# Keep only cryptocurrencies that are on trading
df_clean = df_clean.loc[df_clean['IsTrading'] == True]
df_clean.head()
```

```python
# Keep only cryptocurrencies with a working algorithm
df_clean = df_clean.loc[df_clean['Algorithm']!= 'N/A']
df_clean.head()
```

```python
# Remove the "IsTrading" column
df_clean = df_clean.drop(columns = 'IsTrading')
```

```python
# Removing rows with at least 1 null value
df_clean = df_clean.dropna()
```

```python
# Removing rows with cryptocurrencies without coins mined
df_clean = df_clean.loc[df_clean['TotalCoinsMined'] != 0]
```
